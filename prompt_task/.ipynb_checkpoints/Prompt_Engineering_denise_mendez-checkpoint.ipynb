```python
# 1. Poor Prompt (Baseline)
# Technique: Poorly performing, generic prompt

prompt1 = "Write a Python function to sort a list."
response1 = epam_dial.invoke(prompt1)
print("1. Poor Prompt Output:\n", response1.content)
```

```python
# 2. Add Context
# Technique: Add context for clarity

prompt2 = """
Write a Python function to sort a list of integers in ascending order.
"""
response2 = epam_dial.invoke(prompt2)
print("2. Add Context Output:\n", response2.content)
```

```python
# 3. Add Instruction and Example
# Technique: Instruction + Example

prompt3 = """
Write a Python function to sort a list of integers in ascending order.
For example, given [5, 2, 9], the function should return [2, 5, 9].
"""
response3 = epam_dial.invoke(prompt3)
print("3. Instruction + Example Output:\n", response3.content)
```

```python
# 4. Specify Output Format
# Technique: Specify output format

prompt4 = """
Write a Python function to sort a list of integers in ascending order.
Return only the function code, no explanations.
Example input: [5, 2, 9]
Expected output: [2, 5, 9]
"""
response4 = epam_dial.invoke(prompt4)
print("4. Specify Output Format Output:\n", response4.content)
```

```python
# 5. Add Role and Tone
# Technique: Role + Tone

prompt5 = """
You are a Python programming instructor. Write a clear and well-documented Python function to sort a list of integers in ascending order. Include comments explaining each step.
"""
response5 = epam_dial.invoke(prompt5)
print("5. Role + Tone Output:\n", response5.content)
```

```python
# 6. Chain of Thought (Cognitive Reasoning)
# Technique: Chain of Thought

prompt6 = """
You are a Python programming instructor. Explain step by step how you would write a function to sort a list of integers in ascending order, then provide the code with comments.
"""
response6 = epam_dial.invoke(prompt6)
print("6. Chain of Thought Output:\n", response6.content)
```

```python
# 7. Socratic Prompting
# Technique: Socratic Prompting

prompt7 = """
You are a Socratic mentor. Instead of giving the answer directly, ask me questions to guide me in writing a Python function that sorts a list of integers in ascending order.
"""
response7 = epam_dial.invoke(prompt7)
print("7. Socratic Prompting Output:\n", response7.content)
```

```python
# 8. Self-Reflection / Self-Critique
# Technique: Self-Reflection

prompt8 = """
Write a Python function to sort a list of integers in ascending order. After providing the code, reflect on your solution and suggest any possible improvements or alternative approaches.
"""
response8 = epam_dial.invoke(prompt8)
print("8. Self-Reflection Output:\n", response8.content)
```

```python
# 9. Meta-Prompting
# Technique: Meta-Prompting

meta_prompt = """
You are an expert prompt engineer. Refine the following prompt for clarity, specificity, and effectiveness:
'Write a Python function to sort a list of integers in ascending order. After providing the code, reflect on your solution and suggest any possible improvements or alternative approaches.'
"""
response9 = epam_dial.invoke(meta_prompt)
refined_prompt = response9.content
print("9. Meta-Prompting Output (Refined Prompt):\n", refined_prompt)

# Test the refined prompt
response9b = epam_dial.invoke(refined_prompt)
print("\n9b. Meta-Prompting Test Output:\n", response9b.content)
```

```python
# 10. Validation with a Similar Task
# Technique: Validation and Reflection

prompt10 = """
Write a Python function to sort a list of strings alphabetically. After providing the code, reflect on your solution and suggest any possible improvements or alternative approaches.
"""
response10 = epam_dial.invoke(prompt10)
print("10. Validation with Similar Task Output:\n", response10.content)
```

- The prompt evolves from a vague instruction to a highly effective, context-rich, and self-reflective instruction.
- Meta-prompting and validation ensure the prompt is robust and generalizable.
- You can compare outputs and reflect on which strategies yield the clearest, most useful, or most educational responses.

# Bonus Reflection

This bonus cell allowed me to explore 10 advanced prompt engineering strategies for the same coding task.  
I observed that "Chain of Thought" produced a detailed, step-by-step explanation, while "Socratic Prompting" encouraged me to think and participate in the solution.  
"Meta Prompting" helped refine the prompt itself, and "Self-Critique" showed the model's ability to self-evaluate.  
This exercise demonstrated that the way I ask a question greatly influences the quality and style of the AI's response.

```python
# BONUS: Iterative Prompt Engineering with Multiple Strategies
# This cell demonstrates 10 advanced prompt engineering strategies applied to the same task.
# You can analyze the outputs and reflect on which strategies are most effective.

prompts = [
    # 1. Chain of Thought
    ("Chain of Thought", """
Act as a Python tutor. Think aloud step by step as you write a function to sort a list of integers in ascending order.
"""),
    # 2. Contrastive Chain of Thought
    ("Contrastive Chain of Thought", """
Act as a Python tutor. Consider two different ways to write a function to sort a list of integers in ascending order. Compare the approaches and recommend the best.
"""),
    # 3. Zero-Shot Learning
    ("Zero-Shot Learning", """
Write a Python function to sort a list of integers in ascending order. Do not use any examples.
"""),
    # 4. Few-Shot Learning
    ("Few-Shot Learning", """
Example:
Code:
def reverse_list(lst):
    return lst[::-1]

Analysis:
- The function returns the reversed list.
- Suggestion: Use descriptive variable names.

Now, write a function to sort a list of integers in ascending order and analyze it in the same way.
"""),
    # 5. Emotional Prompting
    ("Emotional Prompting", """
Act as a supportive Python tutor. Kindly write a function to sort a list of integers in ascending order and gently explain your solution.
"""),
    # 6. Ask Me Before You Answer
    ("Ask Me Before You Answer", """
Before writing the function, ask me any clarifying questions you need. Then, write a function to sort a list of integers in ascending order.
"""),
    # 7. Self-Critique
    ("Self-Critique", """
Write a Python function to sort a list of integers in ascending order. After providing the code, critique your own solution and suggest improvements.
"""),
    # 8. Meta Prompting
    ("Meta Prompting", """
You are an expert prompt engineer. Refine the following prompt for clarity, specificity, and effectiveness:
'Write a Python function to sort a list of integers in ascending order.'
"""),
    # 9. Socratic Prompting
    ("Socratic Prompting", """
Act as a Socratic mentor. Instead of giving the answer directly, ask me questions to guide me in writing a function to sort a list of integers in ascending order.
"""),
    # 10. Program of Thought
    ("Program of Thought", """
Follow this sequence:
1. Summarize the task.
2. Write the function.
3. Explain your reasoning.
4. Reflect on possible improvements.

Task: Write a Python function to sort a list of integers in ascending order.
"""),
]

for idx, (name, prompt) in enumerate(prompts, 1):
    print(f"\n--- Iteration {idx}: {name} ---\n")
    response = epam_dial.invoke(prompt)
    print(response.content)
```
