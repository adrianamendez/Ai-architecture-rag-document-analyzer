```python
def bubble_sort(lst):
    """
    Sorts a list in ascending order using the bubble sort algorithm.
    
    Parameters:
    lst (list): The list of elements to be sorted.
    
    Returns:
    list: A new list that is sorted in ascending order.
    """
    n = len(lst)
    # Create a copy of the list to avoid modifying the original
    sorted_lst = lst.copy()
    
    for i in range(n):
        for j in range(0, n-i-1):

            if sorted_lst[j] > sorted_lst[j+1]:
                sorted_lst[j], sorted_lst[j+1] = sorted_lst[j+1], sorted_lst[j]
    
    return sorted_lst
```

### Step-by-step Explanation of Bubble Sort

1. **Define the Function**: We'll create a function named `bubble_sort` which takes a list of integers as an argument.

2. **Loop Through the List**: We'll use a nested loop to compare each pair of adjacent elements in the list. The outer loop will iterate through the entire list.

3. **Compare and Swap**: Inside the inner loop, we will compare each integer with the one next to it. If the current integer is greater than the next one, we will swap them.

4. **Repeat**: We will repeat this process until we make a complete pass through the list without making any swaps, indicating that the list is sorted.

5. **Return the Sorted List**: Finally, we will return the sorted list.

### Code Implementation

Here's how you can implement the Bubble Sort algorithm in Python:

```python
def bubble_sort(arr):
    """
    Sorts a list of integers in ascending order using the Bubble Sort algorithm.

    Parameters:
    arr (list): A list of integers to be sorted.

    Returns:
    list: A new list containing the sorted integers.
    """
    n = len(arr)  # Get the number of elements in the list
    
    # Loop through the entire list
    for i in range(n):
        # Initialize a flag to track if a swap was made
        swapped = False
        
        # Loop through the list from 0 to n-i-1
        for j in range(0, n - i - 1):
            # Compare adjacent elements
            if arr[j] > arr[j + 1]:
                # Swap if the current element is greater than the next element
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True  # Set flag to True if a swap occurred
        
        # If no swaps were made, the list is already sorted
        if not swapped:
            break

    return arr  # Return the sorted list

# Example usage:
numbers = [64, 34, 25, 12, 22, 11, 90]
sorted_numbers = bubble_sort(numbers)
print("Sorted list:", sorted_numbers)
```

### Explanation of the Code:

- **Function Definition**: We define `bubble_sort` that takes a list `arr` as its parameter.
- **Length Calculation**: `n = len(arr)` calculates the length of the list to determine how many times we need to iterate.
- **Outer Loop**: The first `for` loop `(for i in range(n))` runs `n` times for all elements.
- **Inner Loop**: The second `for` loop `(for j in range(0, n - i - 1))` compares each element with the one that follows it; the `-i - 1` ensures we don't compare sorted elements.
- **Swapping Logic**: If an element is greater than the next one, we swap them using a Python tuple assignment.
- **Optimization Check**: The `swapped` flag helps to terminate the loop early if no swaps were made during an iteration, indicating the list is already sorted.
- **Return Statement**: The sorted list is returned at the end.

```python
def sort_integers(int_list):
    """
    Sorts a list of integers in ascending order using the bubble sort algorithm.
    
    Parameters:
    int_list (list): A list of integers to be sorted.
    
    Returns:
    list: A new list containing the sorted integers.
    """
```

### Reflection on Efficiency and Alternatives

**Efficiency:**
- The provided implementation uses the bubble sort algorithm, which is a simple sorting method that repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. This process is repeated until no swaps are needed, indicating that the list is sorted.
- **Time Complexity:** The worst-case and average-case time complexity of bubble sort is \(O(n^2)\), where \(n\) is the number of items being sorted. This occurs when the input list is in reverse order. The best-case time complexity is \(O(n)\), which occurs when the list is already sorted.
- **Space Complexity:** The space complexity is \(O(1)\) because it sorts the list in place and only requires a constant amount of additional space for variables.

**Potential Improvements:**
- Given that bubble sort is not the most efficient algorithm for large datasets, particularly due to its \(O(n^2)\) time complexity, it is usually not recommended for production or performance-sensitive applications. Instead, algorithms like quicksort, mergesort, or heapsort could be used.

**Alternative Sorting Algorithms:**
1. **Quicksort:** Average-case time complexity is \(O(n \log n)\), and space complexity is \(O(\log n)\) due to the call stack.
2. **Mergesort:** Time complexity is consistently \(O(n \log n)\) and has a space complexity of \(O(n)\).
3. **Heapsort:** Also \(O(n \log n)\) for both average and worst cases, with a space complexity of \(O(1)\).

These algorithms are generally more efficient for larger datasets. For even more advanced sorting, Python provides a built-in `sort()` method for lists, which uses Timsort (a hybrid sorting algorithm derived from merge sort and insertion sort), which is highly optimized and suitable for various cases.

In summary, while the bubble sort implementation meets the initial requirements, exploring more efficient sorting methods for larger lists is advisable for real-world applications.

### Approach 1: Using `list.sort()`

The `list.sort()` method sorts a list in place. This means the original list is modified, and it doesn't return a new list.


### How It Works:

### Usage Example:

The `sorted()` function returns a new sorted list from the elements of any iterable.

```python

```

### How It Works:
- This function does not modify the original list; instead, it creates and returns a new sorted list.

### Usage Example:

### Comparison

1. **In-Place vs. New List**:
   - `list.sort()` modifies the original list and does not return anything, which is useful when you want to keep memory usage low and do not need the original list.

   - `list.sort()` is more memory efficient because no additional list is created.
   - `sorted()` may use more memory since it creates a new list.

3. **Readability**:
   - Using `sorted()` makes your intention clear that you are returning a sorted list without altering the original list.
   - Using `list.sort()` might be preferred in cases where the purpose is to modify the list directly and do not require the original order.

### Recommendation

The best approach depends on your specific use case:
- If you need to maintain the original list and require a sorted version, go with `sorted()`.
- If you wish to sort the list in place and don't need the original order, use `list.sort()`.

In general, for code readability and functionality, `sorted()` is often preferred as it allows for flexibility while ensuring that the original data is unchanged. However, if performance and memory constraints are considerations, and you are dealing with large lists where modifying the original is acceptable, then `list.sort()` would be the way to go.

### Implementation of a Bubble Sort Algorithm

Here’s a function that uses the Bubble Sort algorithm to sort a list of integers ascending:

```python
def bubble_sort(arr):
    n = len(arr)  # Get the number of elements in the list
    # Traverse through all elements in the array
    for i in range(n):
        # Last i elements are already sorted, so we can skip them
        for j in range(0, n - i - 1):
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

### How This Works

1. **Understanding the Input**: The function takes a list of integers (`arr`) as its input. 

2. **Outer Loop**: The outer loop runs `n` times, where `n` is the total number of elements in the list. With each iteration of this loop, the largest unsorted element "bubbles" up to its correct position.

3. **Inner Loop**: The inner loop runs from the beginning of the list to the last unsorted element. It compares each element (`arr[j]`) with the next element (`arr[j + 1]`). 

4. **Swapping Elements**: If the current element is greater than the next, they are swapped. This is done using tuple unpacking: `arr[j], arr[j + 1] = arr[j + 1], arr[j]`.

5. **Returning the Sorted List**: After all iterations, the function returns the sorted list.

### Example Usage
